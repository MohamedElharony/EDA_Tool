```python
def load_data(file_path):
    data = pd.read_csv(file_path)
    return data
```

```python
def preprocess_data(data):
    data = data.dropna()
    return data
```

```python
def generate_summaries(data):
    summary = data.describe()
    return summary
```

```python

def visualize_numeric_column(data, column_name):
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column_name], bins=20, kde=True)
    plt.title(f'Distribution of {column_name}')
    plt.show()

def visualize_categorical_column(data, column_name):
    plt.figure(figsize=(10, 6))
    sns.countplot(data[column_name])
    plt.title(f'Count of {column_name}')
    plt.xticks(rotation=45)
    plt.show()

def visualize_relationship(data, x_column, y_column):
    fig = px.scatter(data, x=x_column, y=y_column, title=f'{x_column} vs {y_column}')
    fig.show()
```

```python
def detect_column_types(data):
    column_types = {}
    for column in data.columns:
        if pd.api.types.is_numeric_dtype(data[column]):
            column_types[column] = 'numeric'
        elif pd.api.types.is_categorical_dtype(data[column]):
            column_types[column] = 'categorical'
    return column_types
```

```python
def automated_eda_tool(file_path):
    data = load_data(file_path)
    data = preprocess_data(data)
    summaries = generate_summaries(data)
    column_types = detect_column_types(data)

    for column, col_type in column_types.items():
        if col_type == 'numeric':
            visualize_numeric_column(data, column)
        elif col_type == 'categorical':
            visualize_categorical_column(data, column)

automated_eda_tool('International_Report_Departures.csv')
automated_eda_tool('International_Report_Passengers.csv')
```
